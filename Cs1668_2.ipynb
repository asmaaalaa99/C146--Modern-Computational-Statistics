```
import networkx as nx
import random as random 
alpha = 0.85
N = 100
for i in range (N):
  g = nx.erdos_renyi_graph(30, 0.05, directed=True, seed=123)
  list_of_nodes = list(g.nodes)
  list_of_nodes_with_n = []
  nodes_visited = []
  for n in g.nodes:
    if len(list(g.neighbors(n))) != 0:
       list_of_nodes_with_n.append(n)
  current_node = random.choice(list_of_nodes_with_n)
for _ in range(N):
  current_node = random.choice(list_of_nodes_with_n)
  if (alpha) <= random.random():
    current_node = random.choice(list(g.neighbors(current_node)))
    nodes_visited.append(current_node)
  else:
    current_node = random.choice(list_of_nodes_with_n)
    nodes_visited.append(current_node)
    
```

```
import matplotlib.pyplot as plt
plt.hist(nodes_visited)
```

```
d = {x:nodes_visited.count(x) for x in nodes_visited}
a, b = d.keys(), d.values()
```

```
b = list(b)
freq = []
for _ in b:
  f = _/len(b)
  freq.append(f)
  
```

```
plt.hist(freq)
```

```
network_x = nx.pagerank(g)
network_x_key, network_x_key_val = network_x.keys() , network_x.values()
```

```
difference = []
for m, n in zip(freq, network_x_key_val):
  difference.append(abs(m-n))
```
